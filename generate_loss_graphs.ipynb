```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
runs_folder = "./runs"
```

```python
schedulers = ["linear", "quadratic", "cosine", "sigmoid"]
loss_files = [f"{runs_folder}/{scheduler}/losses.csv" for scheduler in schedulers]
print(loss_files)
```

```python
losses = [pd.read_csv(loss_file) for loss_file in loss_files]
epochs = [i + 1 for i in range(len(losses[0]))]
```

```python
# Plot the losses as subplots
plt.rcParams.update({'font.size': 15})
fig, axs = plt.subplots(1, 4, figsize=(20, 6), constrained_layout=True)
for i in range(len(losses)):
    ax = axs[i]
    ax.set_title(schedulers[i])
    ax.set_xlabel("Epoch")
    ax.set_ylabel("Loss")
    ax.plot(epochs, losses[i]["train_loss"], label="train")
    ax.plot(epochs, losses[i]["val_loss"], label="val")
    ax.legend()
    ax.grid()

# save the plot
plt.savefig("losses.png")
```
